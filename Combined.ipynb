```python
import pandas as pd
```

```python
# Assuming your CSV file is located in the same directory as your notebook
file_path = 'combined_real_estate_weather.csv'
df = pd.read_csv(file_path)
```

```python
# Rename column 'Louisianatitude' to 'latitude'
df.rename(columns={'Louisianatitude': 'latitude'}, inplace=True)
```

```python
# List of columns to keep
columns_to_keep = ['price', 'bedroom_number', 'latitude', 'longitude', 'temperature', 'state']

# Drop columns not in columns_to_keep
columns_to_drop = [col for col in df.columns if col not in columns_to_keep]
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df_cleaned.head()
```

```python
df_cleaned.columns
```

```python
# Define temperature ranges
temperature_ranges = {
    'Cold': (0, 40),
    'Mild': (41, 60),
    'Warm': (61, 80),
    'Hot': (81, 100),
    'Very Hot': (101, float('inf'))  # For temperatures above 100Â°F
}
```

```python
# Function to assign temperature range
def assign_temperature_range(temp):
    for range_name, (min_temp, max_temp) in temperature_ranges.items():
        if min_temp <= temp <= max_temp:
            return range_name
    return 'Unknown'  # Handle cases outside defined ranges
```

```python
# Apply function to create 'temperature_range' column
df_cleaned['temperature_range'] = df_cleaned['temperature'].apply(assign_temperature_range)
```

```python
# Group by 'state' and 'temperature_range' to calculate average prices
grouped_data = df_cleaned.groupby(['state', 'temperature_range'])['price'].mean().reset_index()
```

```python
grouped_data.head()
```

```python
# Drop rows with NaN prices, if any
df = df_cleaned.dropna(subset=['price'])

# Step 4: Calculate average prices by state
state_avg_prices = df_cleaned.groupby('state')['price'].mean().reset_index()
```

```python
state_avg_prices
```

```python
# Step 3: Calculate average temperature by state
state_avg_temperature = df_cleaned.groupby('state')['temperature'].mean().reset_index()
```

```python
state_avg_temperature
```

```python
# Step 3: Merge state_avg_prices and state_avg_temperature on 'state' column
combined_data = pd.merge(state_avg_prices, state_avg_temperature, on='state')
```

```python
# save the combined DataFrame to a new CSV file
combined_data.to_csv('combined_state_avg_data.csv', index=False)
```
